```python
%%sql

CREATE DATABASE IF NOT EXISTS IMDb;

SHOW DATABASES;
```

```python
%%sql

USE IMDb;

SHOW TABLES;
```

```python
%%sql

DESCRIBE Names_;
```

```python
%%sql 

SELECT  t.birth_year, 
        COUNT(t.birth_year) AS counts, 
        SUM(COUNT(t.birth_year)) OVER() AS sum
FROM (SELECT * FROM Names_ LIMIT 1000) t
GROUP BY birth_year
ORDER BY birth_year DESC;
```

```python
%%sql

SELECT t.birth_year, COUNT(t.birth_year) AS counts,
SUM(COUNT(t.birth_year)) OVER() AS sum
FROM (SELECT * FROM Names_ ORDER BY birth_year DESC LIMIT 1000) t
GROUP BY birth_year
ORDER BY birth_year DESC;
```

```python
%%sql

SELECT MIN(birth_year) AS min, MAX(birth_year) AS max
FROM Names_;
```

```python
%%sql

SELECT MIN(birth_year), MAX(birth_year) 
FROM Names_
WHERE birth_year > 1850;
```

```python
%%sql

SELECT birth_year, COUNT(birth_year) AS counts
FROM Names_
WHERE birth_year > 1850
GROUP BY birth_year
ORDER BY birth_year ASC;
```

```python
%%sql

SELECT AVG(birth_year), STD(birth_year)
FROM Names_
WHERE birth_year > 1850;
```

```python
%%sql

DESCRIBE Titles;
```

```python
%%sql

SELECT start_year, COUNT(start_year) AS counts
FROM Titles
GROUP BY start_year
HAVING start_year BETWEEN 1996 AND 2021
ORDER BY start_year DESC;
```

```python
%%sql

SELECT n.name_ AS name, t.primary_title AS title, r.role_ AS role
FROM Had_role r 
INNER JOIN Names_ n
ON r.name_id = n.name_id
INNER JOIN Titles t
ON r.title_id = t.title_id
WHERE r.role_ = "Spider-Man";
```

```python
%%sql 

SELECT t.primary_title AS title
FROM Directors d
INNER JOIN Titles t
ON t.title_id = d.title_id
INNER JOIN Names_ n
ON n.name_id = d.name_id
WHERE n.name_ = "Wes Anderson";
```

```python
%%sql

SELECT n.name_, count(*) AS counts
FROM Had_role r
INNER JOIN Names_ n ON r.name_id = n.name_id
WHERE (
    r.title_id IN (
        SELECT d.title_id
        FROM Directors d
        INNER JOIN Names_ n on d.name_id = n.name_id
        WHERE n.name_ = "Wes Anderson"
    )
)
GROUP BY n.name_
ORDER BY counts DESC
LIMIT 10;
```
